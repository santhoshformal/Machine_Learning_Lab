```python
import pandas as pd
df=pd.read_csv('TvMarketing.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
x_train,x_test,y_train,y_test = train_test_split(df[["TV"]],df["Sales"])
plt.scatter(x_train,y_train)
plt.plot()
plt.title("graphical representation of train data")
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X=x_train, y=y_train)
# plt.scatter(x_train,y_train)
# plt.show()
print(model.coef_, model.intercept_)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train),color='red')
plt.title("Linear Regression Model for TV sales")
plt.xlabel("TV")
plt.ylabel("sales")
```

```python
model.predict([[234,]])
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,model.predict(x_test),color='red')
plt.title("Linear Regression Model(test) for TV sales")
plt.xlabel("TV")
plt.ylabel("sales")
```

```python
import numpy as np
valid=np.array([[112],[232]])
predicted_values=model.predict(valid)
predicted_values
```
